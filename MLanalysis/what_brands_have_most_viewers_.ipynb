````
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('youtube_cleaned.csv')

# Group by channel title and sum the view counts
channel_views = df.groupby('channel_title')['view_count'].sum().sort_values(ascending=False)

# Get the top 10 channels with the most views
top_10_channels = channel_views.head(10)

print("Top 10 channels by view count:")
print(top_10_channels)

# Create a bar chart for the top 10 channels
plt.figure(figsize=(12, 8))
top_10_channels.sort_values(ascending=True).plot(kind='barh', color='coral')
plt.xlabel('Total View Count (in hundred millions)')
plt.ylabel('Channel')
plt.title('Top 10 Channels by Total View Count')
plt.tight_layout()

# Save the plot
plt.savefig('top_channels_by_views.png')

print("\nGenerated a bar chart of the top 10 channels by view count and saved it as top_channels_by_views.png")
````
