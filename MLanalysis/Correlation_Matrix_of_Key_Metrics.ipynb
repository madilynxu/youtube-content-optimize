````
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('youtube_cleaned.csv')

# --- Correlation Analysis ---
# Select key numerical columns for correlation analysis
correlation_df = df[['view_count', 'like_count', 'comment_count', 'subscriber_count', 'duration_seconds']]

# Calculate the correlation matrix
correlation_matrix = correlation_df.corr()

print("Correlation Matrix:")
print(correlation_matrix)

# --- Visualization ---
# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key YouTube Metrics')
plt.tight_layout()

# Save the plot
plt.savefig('correlation_heatmap.png')

print("\nGenerated a heatmap of the correlation matrix and saved it as correlation_heatmap.png")
````

### **Key Takeaways**

Here are some of the main takeaways from this analysis:

  * **Views, Likes, and Subscribers are Strongly Linked:** There is a very strong positive correlation between `view_count`, `like_count`, and `subscriber_count`. This is the most important relationship in the data. It means that videos with high view counts also tend to have high like counts, and they are typically on channels with more subscribers. This suggests a virtuous cycle: popular videos drive likes and subscriber growth, and larger channels get more views on their new videos.

  * **Video Length Doesn't Directly Drive Engagement:** Interestingly, `duration_seconds` has a very weak correlation with views, likes, comments, and subscribers. This implies that **video length is not a determining factor** in a video's success, at least in this dataset. A short video can be just as successful as a long one. The focus should be on the quality and appeal of the content rather than its length.

  * **Comments Show a Different Engagement Pattern:** While `comment_count` is positively correlated with views and likes, the relationship isn't as strong as the one between views and likes themselves. This suggests that getting a comment requires a higher level of engagement from a viewer than just liking a video.

  * **Content is King:** As seen from the tag analysis, the most popular content falls into specific niches like **gaming** (`Minecraft`, `Roblox`), **combat sports** (`boxing`, `UFC`, `WWE`), and mainstream **entertainment** from major brands like Netflix and Disney. The most successful channel in your dataset, **MrBeast**, is known for high-production, high-stakes challenge videos, a unique and popular content formula. This highlights that success on YouTube is highly dependent on the type and quality of content you produce.
